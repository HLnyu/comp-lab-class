```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# data path
colvar_1_path = '../Inputs/exercise2/exercise2_sigma55height1.2bf10.colvar.txt'
hills_1_path = '../Inputs/exercise2/exercise2_sigma55height1.2bf10.hills.txt'
grid_1_path = '../Inputs/exercise2/exercise2_sigma55height1.2bf10.dat'
# load data
colvar_1 = np.loadtxt(colvar_1_path)
hills_1 = np.loadtxt(hills_1_path)
grid_1 = np.loadtxt(grid_1_path)
# check data
print('shape of colvar_1:', colvar_1.shape)
print('shape of hills_1:', hills_1.shape)
print('shape of grid_1:', grid_1.shape)
```

```python
# figure 3 - time evolution of psi during first 2 ns
fig=plt.figure(3)

# data processing, load time in ns
time_1 = colvar_1[:,0]/1000
phi_1 = colvar_1[:,1]

# plot phi over time, scatter, first 100000 frame(2ns), colorblind friendly
plt.scatter(time_1[:100000], phi_1[:100000], s=1, c='tab:blue')

# set axis range
plt.xlim(-0.05,2.05)
plt.ylim(-np.pi, np.pi)

# labels, font size 12
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel(r'$\phi$ (rad)', fontsize=12)

# two lines of title
plt.text(x=0.5, y=0.88, s="Time evolution of "r"$\phi$"" (0-2 ns)", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s= "ADP, 10 ns metad, biasfactor10_pace500_height12_sigma0.55", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)

# save figure as pdf in ../Figures
fname = '../Figures/Time_evolution_of_phi[0-2ns].pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```

```python
fig,axs=plt.subplots(1,3,layout='constrained', figsize=(12,4))

# data processing, load time in ns
time_1 = colvar_1[:,0]/1000
phi_1 = colvar_1[:,1]
psi_1 = colvar_1[:,2]
bias_1 = colvar_1[:,3]

# set colorblind-friendly palatte 
colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB']

# plot phi over time, scatter, colorblind friendly
axs[0].scatter(time_1, phi_1, s=1, c=colors[0])
# set axis range
axs[0].set_xlim(-0.05,10.05)
axs[0].set_ylim(-np.pi, np.pi)
# labels, font size 12
axs[0].set_xlabel('Time (ns)', fontsize=12)
axs[0].set_ylabel(r'$\phi$ (rad)', fontsize=12)
# title
axs[0].set_title("Time evolution of "r"$\phi$", fontsize=12)

# plot psi over time, scatter, colorblind friendly
axs[1].scatter(time_1, psi_1, s=0.1, c=colors[1])
# set axis range
axs[1].set_xlim(-0.05,10.05)
axs[1].set_ylim(-np.pi, np.pi)
# labels, font size 12
axs[1].set_xlabel('Time (ns)', fontsize=12)
axs[1].set_ylabel(r'$\psi$ (rad)', fontsize=12)
# title
axs[1].set_title("Time evolution of "r"$\psi$", fontsize=12)

# plot bias over time, scatter, colorblind friendly
axs[2].scatter(time_1, bias_1, s=0.1, c=colors[2])
# set axis range
axs[2].set_xlim(-0.05,10.05)
# labels, font size 12
axs[2].set_xlabel('Time (ns)', fontsize=12)
axs[2].set_ylabel('Bias (kJ/mol)', fontsize=12)
# title
axs[2].set_title("Time evolution of bias", fontsize=12)

fig.show()

# show bias distribution over phi & psi
fig,axs=plt.subplots(1,2,layout='constrained', figsize=(8,4))

# plot bias over phi, scatter, colorblind friendly
axs[0].scatter(phi_1, bias_1, s=0.1, c=colors[0])
# set axis range
axs[0].set_xlim(-np.pi, np.pi)
axs[0].set_xlim(-np.pi, np.pi)
# labels, font size 12
axs[0].set_xlabel(r'$\phi$ (rad)', fontsize=12)
axs[0].set_ylabel('Bias (kJ/mol)', fontsize=12)
# title
axs[0].set_title("Bias over "r"$\phi$", fontsize=12)

# plot bias over psi, scatter, colorblind friendly
axs[1].scatter(psi_1, bias_1, s=0.1, c=colors[1])
# set axis range
axs[1].set_xlim(-np.pi, np.pi)
axs[1].set_xlim(-np.pi, np.pi)
# labels, font size 12
axs[1].set_xlabel(r'$\psi$ (rad)', fontsize=12)
axs[1].set_ylabel('Bias (kJ/mol)', fontsize=12)
# title
axs[1].set_title("Bias over "r"$\psi$", fontsize=12)
```

```python
# figure 4 - time evolution of Gaussian height
fig=plt.figure(4)

# data processing, load time in ns
time_1 = hills_1[:,0]/1000
phi_1 = hills_1[:,1]
height_1 = hills_1[:,3]

# set colorblind-friendly palatte 
colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB']

# plot height over time, scatter, colorblind friendly
plt.scatter(time_1, height_1, s=2, c=colors[0])

# set axis range
plt.xlim(-0.08,10.05)
plt.ylim(-0.01, 1.42)

# labels, font size 12
plt.xlabel('Time (ns)', fontsize=12)
plt.ylabel('Gaussian height (kJ/mol)', fontsize=12)

# two lines of title
plt.text(x=0.5, y=0.88, s="Time evolution of Gaussian height", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.82, s= "ADP, 10 ns metad, biasfactor10_pace500_height12_sigma0.55", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.3)

fname = '../Figures/Time_evolution_of_Gaussian_height.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```

```python
# figure 5 - bias landscape & free energy landscape
fig,axs=plt.subplots(1,2,layout=None, figsize=(12,5))
# data processing, load time in ns
phi_1 = grid_1[:,0]
bias_1 = grid_1[:,1]
derivatives_phi_1 = grid_1[:,2]
biasfactor_1 = 10
free_energy_1 = bias_1 * (biasfactor_1/(1-biasfactor_1))
free_energy_derivatives_phi_1 = derivatives_phi_1 * (biasfactor_1/(1-biasfactor_1))

# set colorblind-friendly palatte
colors = ['#0077BB', '#33BBEE', '#009988', '#EE7733', '#CC3311', '#EE3377', '#BBBBBB']

# plot minus bias potential over phi, colorblind friendly, line width 2
axs[0].plot(phi_1, -bias_1, c=colors[0], lw=2, label='minus bias potential ''-V')
# set axis range
axs[0].set_xlim(-np.pi, np.pi)
# labels, font size 12
axs[0].set_xlabel(r'$\phi$ (rad)', fontsize=12)
axs[0].set_ylabel('Minus Bias Potential (kJ/mol)', fontsize=12)
# title
axs[0].set_title("Minus bias potential landscape", fontsize=12)
# plot derivatives of minus bias potential over phi, colorblind friendly, line width 1, dashed, right y axis
ax0_2=axs[0].twinx()
ax0_2.plot(phi_1, -derivatives_phi_1, c=colors[1], lw=1, ls='--', label='minus derivative 'r'$-\frac{\partial V}{\partial \phi}$')
# labels, font size 12
ax0_2.set_ylabel('Derivative (kJ/mol'r'$\bullet$''rad))', fontsize=12)
# combine legend for twinx
axs[0].legend(loc='upper left', bbox_to_anchor=(0,1), bbox_transform=axs[0].transAxes, fontsize=12, frameon=False)
ax0_2.legend(loc='upper left', bbox_to_anchor=(-0.003,0.95), bbox_transform=axs[0].transAxes, fontsize=12, frameon=False)

# plot free energy over phi, colorblind friendly, line width 2
axs[1].plot(phi_1, free_energy_1, c=colors[0], lw=2, label='free energy 'r'$F$')
# set axis range
axs[1].set_xlim(-np.pi, np.pi)
# labels, font size 12
axs[1].set_xlabel(r'$\phi$ (rad)', fontsize=12)
axs[1].set_ylabel('Free Energy (kJ/mol)', fontsize=12)
# title
axs[1].set_title("Free energy landscape", fontsize=12)
# plot derivatives of free energy over phi, colorblind friendly, line width 1, dashed, right y axis
ax1_2=axs[1].twinx()
ax1_2.plot(phi_1, free_energy_derivatives_phi_1, c=colors[1], lw=1, ls='--', label='derivative 'r'$\frac{\partial F}{\partial \phi}$')
# labels, font size 12
ax1_2.set_ylabel('Derivative (kJ/mol'r'$\bullet$''rad))', fontsize=12)
# combine legend for twinx
axs[1].legend(loc='upper left', bbox_to_anchor=(0,1.01), bbox_transform=axs[1].transAxes, fontsize=12, frameon=False)
ax1_2.legend(loc='upper left', bbox_to_anchor=(-0.003,0.94), bbox_transform=axs[1].transAxes, fontsize=12, frameon=False)

# title
plt.text(x=0.5, y=0.95, s="Bias potential & free energy landscape", fontsize=18, ha="center", transform=fig.transFigure)
plt.text(x=0.5, y=0.9, s= "ADP, 10 ns metad, biasfactor10_pace500_height12_sigma0.55", fontsize=11, ha="center", transform=fig.transFigure)
plt.subplots_adjust(top=0.8, wspace=0.4)

# save figure as pdf in ../Figures
fname = '../Figures/[metad_10ns_topolA]Energy_landscape.pdf'
plt.savefig(fname, format='pdf', dpi=300, bbox_inches='tight')
```
