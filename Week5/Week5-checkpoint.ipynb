```python
# Week 5 task
import mdtraj as md
import matplotlib.pyplot as plt
import numpy as np
```

```python
#load file
trr_file = "../Inputs/adp.trr"
gro_file = "../Inputs/adp.gro"
traj = md.load (trr_file, top = gro_file)
```

```python
phi =md.compute_phi(traj)
psi= md.compute_psi(traj)
```

```python
# plot phi vs time
plt.plot(traj.time / 1000, phi[1])
plt.xlabel('Time (ns)')
plt.ylabel('Phi (rad)')
plt.title('Phi vs Time')
plt.show()
```

```python
# plot psi vs time
plt.plot(traj.time / 1000, psi[1])
plt.xlabel('Time (ns)')
plt.ylabel('Psi (rad)')
plt.title('Psi vs Time')
plt.show()
```

```python
phi[1].max()
```

```python
H, xedges, yedges = np.histogram2d(phi[1].flatten(),psi[1].flatten(),bins = 100)
H = -np.log(H.T)
H = H - np.min(H)
F = -np.log(H)
F -= F.min()
Fmin, Fmax = 0, 10
levels = np.linspace(Fmin, Fmax, 11)
x = 0.5*(xedges[1:] + xedges[:-1])
y = 0.5*(yedges[1:] + yedges[:-1])
plt.imshow(H, interpolation= 'nearest', origin="lower", extent=[xedges[0], xedges[-1], yedges[0], yedges[-1]])
plt.xlabel ('Phi(rad)')
plt.ylabel ('Psi(rad)')
plt.contourf(x, y, F.T, levels=levels)
# colorbar, 0 to 8, fontsize = 12, only show ticks at 0, 2, 4, 6, 8, 10
cbar = plt.colorbar(ticks=[0, 2, 4, 6, 8, 10])
cbar.ax.tick_params(labelsize=12)
# add labels to the colorbar, fontsize = 12
cbar.set_label('Free energy (kT)', fontsize=12)
plt.show()
```

```python
#Replica analysis
# read in xvg files from 20ns_T300
replica_index = np.loadtxt('../Inputs/20ns_T300//replica_index.xvg')
replica_temp = np.loadtxt('../Inputs/20ns_T300/replica_temp.xvg')
```

```python
# plot the replica temperature every 500 steps
idx_every_500 = np.arange(0, len(replica_temp), 500)
replica_temp_every_500 = replica_temp[idx_every_500, 1]
# change the replica index into temperature
replica_temp_every_500[replica_temp_every_500 == 0] = 300
replica_temp_every_500[replica_temp_every_500 == 1] = 400
replica_temp_every_500[replica_temp_every_500 == 2] = 500
replica_temp_every_500[replica_temp_every_500 == 3] = 600
plt.plot(replica_temp[idx_every_500,0], replica_temp_every_500)
plt.xlabel('Time (ps)')
plt.ylabel('Temperature (K)')
plt.title('Replica Temperature')
plt.show()
```

```python
# calculate the exchange probability
exchange_prob = np.zeros_like(replica_temp[:, 1])
for i in range(1, len(exchange_prob)):
    if replica_index[i, 1] != replica_index[i-1, 1]:
        exchange_prob[i] = 1
    else:
        exchange_prob[i] = 0
print('exchange probability: ', np.sum(exchange_prob)/len(exchange_prob))
```

```python
# plot the histogram of potential energies of each replica
replica_energy = []
for i in [300, 400, 500, 600]:
    potential_energy = np.loadtxt('../Inputs/20ns_T{}/replica_energy.xvg'.format(i), skiprows=24)
    replica_energy.append(potential_energy)
    plt.hist(potential_energy[:,1], bins=100, alpha=0.5, label='T={}'.format(i), density=True)
plt.xlabel('Potential Energy (kJ/mol)')
plt.ylabel('Probability')
plt.title('Potential Energy Histogram')
plt.legend()
plt.show()
```
