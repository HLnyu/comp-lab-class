```python
import numpy as np
import matplotlib.pyplot as plt
import mdtraj as md
```

```python
DATA = np.loadtxt("../Data/1hz3_T310.run.25000000.energy.xvg", skiprows=29)
```

```python
plt.plot(DATA[:,0]*0.001,DATA[:,1])
plt.xlabel(" Time (ns)")
plt.ylabel(" Potential (kJ/mol) ")
plt.title(" Time (ns) VS Potential (kJ/mol)")
```

```python
plt.plot(DATA[:,0]*0.001,DATA[:,2])
plt.xlabel(" Time (ns)")
plt.ylabel(" Kinetic En. (kJ/mol) ")
plt.title(" Time (ns) VS Kinetic En. (kJ/mol)")
```

```python
plt.plot(DATA[:,0]*0.001,DATA[:,3])
plt.xlabel(" Time (ns)")
plt.ylabel(" Total Energy (kJ/mol) ")
plt.title(" Time (ns) VS Total Energy (kJ/mol)")
```

```python
plt.plot(DATA[:,0]*0.001,DATA[:,4])
plt.xlabel(" Time (ns)")
plt.ylabel(" Temperature (K) ")
plt.title(" Time (ns) VS Temperature (K)")
```

```python
plt.plot(DATA[:,0]*1000,DATA[:,5])
plt.xlabel(" Time (ns)")
plt.ylabel(" Pressure (bar) ")
plt.title(" Time (ns) VS  Pressure (bar)")
```

```python
plt.plot(DATA[:,0]*0.001,DATA[:,6])
plt.xlabel(" Time (ns)")
plt.ylabel(" Volume (nm^3) ")
plt.title(" Time (ns) VS Volume (nm^3)")
```

```python
# create figure and axis objects with subplots()
fig,ax = plt.subplots()
# make a plot
ax.plot(DATA[:,0]*0.001,
        DATA[:,4],
        color="red", 
        marker="o")
# set x-axis label
ax.set_xlabel("Time (ns))", fontsize = 14)
# set y-axis label
ax.set_ylabel("Temperature (K)",
              color="red",
              fontsize=14)
# twin object for two different y-axis on the sample plot
ax2=ax.twinx()
# make a plot with different y-axis using second axis object
ax2.plot(DATA[:,0] *0.001,DATA[:,2]/100,color="blue",marker="o", linestyle=":")
ax2.set_ylabel("Kinetic En. (100 kJ/mol)",color="blue",fontsize=14)
plt.title(" Time (ns) VS Temperature (K) & Kinetic En. (100 kJ/mol)")
plt.show()
```
